```python
import re
import pandas as pd
import json
import numpy as np
```

```python
with open("/content/labeled_data.json", "r", encoding="utf-8") as json_file:
    train = json.load(json_file)
print(train['data'][:5])
```

```python
def format_prices(data):
    formatted_data = []

    for item in data:
        sentence = item['sentence']
        labels = item['labels']

        # 숫자 패턴에 일치하는 것을 찾아서 쉼표를 추가
        sentence = re.sub(r'(\d)(?=(\d{3})+(?!\d))', r'\1,', sentence)

        # labels 내의 start와 end 값을 업데이트
        for label in labels:
            if label['entity'] != '가격':
              label['start'] += 1
            label['end'] += 1

        formatted_data.append({'sentence': sentence, 'labels': labels})

    return formatted_data

# 가격 데이터 형식 변경
formatted_data = format_prices(train['data'])

# 결과 출력
formatted_data

# Convert the labeled data to JSON format
labeled_json = json.dumps({'data': formatted_data}, ensure_ascii=False, indent=2)

# Writing the JSON to a file
output_file_path = '/content/labeled_data.json'
with open(output_file_path, 'w', encoding='utf-8') as file:
    file.write(labeled_json)
```

```python
def format_prices_dol(data):
    formatted_data = []

    for item in data:
        sentence = item['sentence']
        labels = item['labels']

        # 숫자 패턴에 일치하는 것을 찾아서 쉼표를 추가
        sentence = '₩' + sentence

        # labels 내의 start와 end 값을 업데이트
        for label in labels:
            if label['entity'] != '가격':
              label['start'] += 1
            label['end'] += 1

        formatted_data.append({'sentence': sentence, 'labels': labels})

    return formatted_data

# 가격 데이터 형식 변경
formatted_data2 = format_prices_dol(train['data'])

# Convert the labeled data to JSON format
labeled_json2 = json.dumps({'data': formatted_data2}, ensure_ascii=False, indent=2)

# Writing the JSON to a file
output_file_path = '/content/augment2.json'
with open(output_file_path, 'w', encoding='utf-8') as file:
    file.write(labeled_json2)
```

```python
def format_prices_won(data):
    formatted_data = []

    for item in data:
        sentence = item['sentence']
        labels = item['labels']

        # 숫자 패턴에 일치하는 것을 찾아서 쉼표를 추가
        sentence = re.sub(r'(\d{1,3}(?:,\d{3})*)(?=\s|$)', lambda x: f"{int(x.group(0)):,}원", sentence)

        # labels 내의 start와 end 값을 업데이트
        for label in labels:
            if label['entity'] != '가격':
              label['start'] += 1
            label['end'] += 1

        formatted_data.append({'sentence': sentence, 'labels': labels})

    return formatted_data

# 가격 데이터 형식 변경
formatted_data3 = format_prices_won(train['data'])

# Convert the labeled data to JSON format
labeled_json3 = json.dumps({'data': formatted_data3}, ensure_ascii=False, indent=2)

# Writing the JSON to a file
output_file_path = '/content/augment3.json'
with open(output_file_path, 'w', encoding='utf-8') as file:
    file.write(labeled_json3)
```
